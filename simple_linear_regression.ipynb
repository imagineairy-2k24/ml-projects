```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# X = study hours, y = exam scores
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([40, 50, 60, 65, 80])
```

```python
# Create the model
model = LinearRegression()

# Train the model on the data
model.fit(X, y)
```

```python
# Predict score for 6 hours of study
prediction = model.predict([[6]])
print(f"Predicted score for 6 hours of study: {prediction[0]:.2f}")
```

```python
# Plot the data and the regression line
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, model.predict(X), color='red', label='Regression line')
plt.scatter([[6]], prediction, color='green', label='Predicted (6 hrs)')
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Linear Regression: Study Hours vs Exam Score")
plt.legend()
plt.grid(True)
plt.show()
```
