```python
print("🌍 Welcome to the Smart Travel Suggestion Bot!\n")

# Collect user inputs
budget = input("💸 Enter your budget (low / medium / high): ").lower()
season = input("☀️ Enter the season you plan to travel (summer / winter / monsoon): ").lower()
interest = input("🎯 What kind of experience do you prefer? (adventure / relaxation / culture): ").lower()
```

```python
# Rule-Based Decision Tree
if budget == "low":
    if season == "summer" and interest == "adventure":
        destination = "Rishikesh (River Rafting)"
    elif season == "winter" and interest == "culture":
        destination = "Jaipur (Heritage Forts)"
    elif season == "monsoon" and interest == "relaxation":
        destination = "Munnar (Hill Station, Kerala)"
    else:
        destination = "Try exploring local weekend destinations!"

elif budget == "medium":
    if season == "summer" and interest == "culture":
        destination = "Hampi (Historical Ruins)"
    elif season == "winter" and interest == "relaxation":
        destination = "Manali (Snow Retreat)"
    elif season == "monsoon" and interest == "adventure":
        destination = "Cherrapunji (Trekking & Caves)"
    else:
        destination = "Explore national gems like Udaipur or Darjeeling!"

elif budget == "high":
    if season == "summer" and interest == "relaxation":
        destination = "Bali, Indonesia (Luxury Beach Resorts)"
    elif season == "winter" and interest == "adventure":
        destination = "Switzerland (Ski & Snowboarding)"
    elif season == "monsoon" and interest == "culture":
        destination = "Kyoto, Japan (Temple Trails)"
    else:
        destination = "Premium international tour packages await!"

else:
    destination = "Sorry, we couldn't understand your budget preference."
```

```python
# Final Output
print("\n🎉 Based on your preferences...")
print(f"✅ Suggested destination for your trip: {destination}")
```
